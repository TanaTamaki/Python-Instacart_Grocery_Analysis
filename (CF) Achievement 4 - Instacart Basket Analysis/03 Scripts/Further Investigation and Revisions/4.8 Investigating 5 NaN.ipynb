### Step 2 - Find the aggregated mean of the 'order_number' column grouped by 'department_id' for the entire dataframe.

```python
# Grouping data by 'department_id'.
# Applying .agg() function to the 'order_number' column to find the mean number of orders placed.
# Finding the average number of orders per user for each department ID. 

df_ords_prods_merge.groupby('department_id').agg({'order_number':['mean']})
```

### Step 3 - Analyze the result. How do the results for the entire dataframe differ from those of the subset?

This code gave us the average number of orders per user for each department id, to know which department(s) sell better.

When comparing the means for the department ids of the first million rows (subset) vs. the entire dataframe, the means appear to be similar.

Except for a few department ids, the means do not appear much different, but a t-test could be run to see if the means are significantly different.

For both the entire dataframe and the subset, the mean of department 21 was the highest (25.53 for subset).

The department ids with the biggest differences is department 10 and 21. The mean of department 10 for the entire dataframe is higher by about 1.5 points compared to the subset mean (18.68). While the mean of department 21 in the entire dataframe has over a 2 point difference compared to the subset mean (25.53) and is lower.

I think overall, it's surprising that the small subset is almost representative of the entire dataframe. There are differences between the two sets of means, but it actually gave a good sense of where the entire data is (at least when calculating the means for this particular column).

### Step 4 - Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform and loc functions.¶

```python
# Creating a new column ('max_order').
# Grouping user IDs and selecting column 'order_number' to transform to find the max order number for each user (for every row).
# Putting the results of the transform function into the new column.

df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

### Step 5 - Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category.

```python
# Trying to determine whether the prices of the products purchased by loyal customers differ from those purchased by regular or new customers. 

df_ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

Note:

- Loyal customers appear to purchase more lower-cost items on average, which could be due to familiarity (knowing when items go on sale, using discounts, etc.)
- Regular customers are in the middle with an average of about $12.50 per item.

### Step 7 - Create an order frequency flag that marks the regularity of a user's ordering behavior according to the median in the 'days_since_prior_order' column.

```python
# Frequent vs. Non-frequent customers and their ordering behavior. 
# Finding the median in 'days_since_prior_order' column for each user and grouping by user.
# Creating a column to put those results into.

df_ords_prods_merge['median_days_between_orders'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
# Checking dataframe.

df_ords_prods_merge.head()
```

```python
# Creating frequency flags based on the median column created and the conditions listed. 

df_ords_prods_merge.loc[df_ords_prods_merge['median_days_between_orders'] > 20, 'ordering_behavior'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_days_between_orders'] > 10) & (df_ords_prods_merge ['median_days_between_orders'] <= 20), 'ordering_behavior'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_between_orders'] <= 10, 'ordering_behavior'] = 'Frequent customer'
```

```python
# Checking count of flags.

df_ords_prods_merge['ordering_behavior'].value_counts(dropna = False)
```
