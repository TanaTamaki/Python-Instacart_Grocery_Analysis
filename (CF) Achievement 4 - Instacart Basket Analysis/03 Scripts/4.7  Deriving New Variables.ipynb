## User-Defined Functions

```python
# Defining a function 

def price_label(row):

    if row['prices'] <= 5:
        return 'Low-range Product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range Product'
    elif row['prices'] > 15:
        return 'High range product'
    else: return np.nan
```

This function called 'price_label' will apply a string to every row within the dataframe, designating it as low-, mid-, or high-range based on its price.

```python
# Applying the function
## Making a new column ('price_range') and making it so that the function output (labels) will be put into this new column 

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Checking the values (labels) in the new column

df['price_range'].value_counts(dropna = False)
```

```python
# Checking the max price in the 'price' column

df['prices'].max()
```

```python
# Creating conditions
## Applies conditional logic of an if-statement to a function without explicitly creating an if-else construct
## Use different cells for each condition

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

## If-Statements with For-Loops

```python
# Summarizing how busy each day of the week is
## First - finding the busiest day of the week for orders, by finding frequency of 'orders_day_of_week' column

df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

Busiest day of the week with the most orders is (0), meaning Saturday.
Slowest day of the week with the fewest orders if (4), meaning Wednesday.

```python
# Making the for-loop to look through every row in the 'ordrers_day_of_week' column and labeling 0 and 4.

result = []

for value in df_ords_prods_merge['orders_day_of_week']:
    if value == 0:
        result.append("Busiest day")
    elif value == 4:
        result.append("Least busy")
    else:
        result.append("Regular busy")
```

```python
result
```

```python
# Making/adding a column to df for these labels to go - summary column for the busiets day of the week 

df_ords_prods_merge['busiest_day'] = result
```

```python
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

### Step 2 - Update "Busiest day" column. 

```python
# Creating a new column 'Busiest days' to include the two busiest and two slowest days of the week. 

## Finding busiest days again

df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

- Busiest days with the most orders are Saturday and Sunday (0, 1).
- Slowest days with the least amount of orders are Tuesday and Wednesday (3, 4).

```python
# Creating for-loop.

day_label = []

for value in df_ords_prods_merge['orders_day_of_week']:
    if value in [0,1]:
       day_label.append("Busiest days")
    elif value in [3,4]:
        day_label.append("Slowest days")
    else:
        day_label.append("Regularly busy")      
```

```python
day_label
```

```python
#Creating/adding new column to df containing these labels.

df_ords_prods_merge['busiest_days'] = day_label
```

```python
# Checking counts.

df_ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

Observations:
- The number of "Busiest days" is equal to the sum of the frequency counts for 0 and 1 (output 29).

### Step 4 - Identify the busiest hours of the day. Create a new column containing these labels called "busiest_period_of_day".

```python
# Finding when most orders are placed during which hour.

df_ords_prods_merge_new['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Creating three distinct periods ("Most orders", "Average orders", and "Fewest orders"), rather than go by hour.
## Will breakdown hours into thirds (8 hours each).

hour_label = []

for value in df_ords_prods_merge_new['order_hour_of_day']:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        hour_label.append("Most orders")
    elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
        hour_label.append("Fewest orders")
    else:
        hour_label.append("Average orders")
```

```python
# Making a new column for these labels and adding to dateframe. 

df_ords_prods_merge_new['busiest_period_of_day'] = hour_label
```
